### MISS NEAR MISS TABLES:

```python
# 
#  TOP-K  MISS / NEAR-MISS  CONTRIBUTION  (one pred + one GT array per model)
# 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 
#  CONSTANTS  (edit if desired)
# 
TOP_K   = 5
NEAR_TH = 0.50          # KS  0.50    miss
GOOD_TH = 0.85          # 0.500.85    near-miss ;  0.85  good
SIGMA   = 0.025         # OKS sigma

CONF_DIR = Path(
    "/Users/marcusschou/Desktop/SDU/data_science/kurser_4/"
    "Speciale_Data_Science/INFERENCE_EVAL/eval_folder/latex_code/test_subset"
)
CONF_DIR.mkdir(parents=True, exist_ok=True)

# 
#  YOUR MODEL DATA  (each array already frame-aligned, shape (F, K, 4))
#  Replace the variables on the right-hand side with your actual arrays.
# 
models_and_arrays = [
    ("SLEAP - Top-down model",pred_slp_ci_td, gt_slp_ci_td),
    ("SLEAP - Single-instance model", pred_slp_si, gt_slp_si),
    ("DLC - Top-down model", pred_dlc_td, gt_dlc_td),
    ("DLC - Bottom-up model", pred_dlc_bu, gt_dlc_bu),
    ("LP - Bottom-up model", pred_lp_bu, gt_lp_bu),
    ("LP - TCN model", pred_lp_c, gt_lp_c),
    ("LP - TCN LCW model", pred_lp_c_new, gt_lp_c_new),
]

# key-point names in correct order
keypoint_names = pd.read_csv(
    "/Users/marcusschou/Desktop/SDU/data_science/kurser_4/"
    "Speciale_Data_Science/data/skeletons/natskeleton_noposs.csv"
)["name"].tolist()

# 
#  HELPER FUNCTIONS  (all defined here)
# 
def add_visibility(gt_xy):
    """Append visibility flag (1 if both coords finite) as 3rd channel."""
    vis = (~np.isnan(gt_xy[..., 0]) & ~np.isnan(gt_xy[..., 1])).astype(float)
    return np.dstack([gt_xy, vis])

def per_frame_area(coords):
    """BBox area per frame, ignoring NaNs. coords: (F, K, 2)."""
    x_min = np.nanmin(coords[..., 0], axis=1)
    x_max = np.nanmax(coords[..., 0], axis=1)
    y_min = np.nanmin(coords[..., 1], axis=1)
    y_max = np.nanmax(coords[..., 1], axis=1)
    return (x_max - x_min) * (y_max - y_min)          # (F,)

def compute_ks_matrix(pred, gt, sigma=SIGMA):
    F, K = pred.shape[:2]
    ks_mat = np.full((F, K), np.nan, dtype=np.float32)
    gt_vis = add_visibility(gt[..., :2])
    pred_xy = pred[..., :2]

    for f in range(F):
        gt_frame = gt_vis[f]
        pred_frame = pred_xy[f]
        vis_flags = gt_frame[:, 2] == 1
        pred_valid = ~np.isnan(pred_frame[:, 0]) & ~np.isnan(pred_frame[:, 1])

        for k in range(K):
            if not vis_flags[k]:
                continue
            if not pred_valid[k]:
                ks_mat[f, k] = 0
                continue

            if k == K - 1:
                # Worm: use full bbox of all visible keypoints
                area_kps = gt_frame[vis_flags, :2]
            else:
                # Others: exclude worm from bbox
                visible_mask = vis_flags.copy()
                visible_mask[K - 1] = False
                area_kps = gt_frame[visible_mask, :2]

            area = per_frame_area(area_kps[None])[0]
            if area == 0:
                continue

            s = np.sqrt(area)
            d = np.linalg.norm(pred_frame[k] - gt_frame[k, :2])
            ks = np.exp(-d**2 / (2 * s**2 * sigma**2))
            ks_mat[f, k] = ks

    return ks_mat

def contribution_per_kp(ks_mat, near_th=NEAR_TH, good_th=GOOD_TH):
    valid      = ~np.isnan(ks_mat)              # every non-NaN sample
    miss_mask  = (ks_mat <= near_th) & valid
    near_mask  = (ks_mat > near_th) & (ks_mat < good_th) & valid

    miss_total = miss_mask.sum()
    near_total = near_mask.sum()

    miss_share = 100 * miss_mask.sum(0) / miss_total if miss_total else np.zeros(ks_mat.shape[1])
    near_share = 100 * near_mask.sum(0) / near_total if near_total else np.zeros(ks_mat.shape[1])
    return miss_share, near_share

# ---- tiny helpers for LaTeX / PNG output (same style you used before) -------
def cell_latex(cell):
    if pd.isna(cell):
        return ""
    s = str(cell).replace("%", r"\%")
    if "(" in s and ")" in s:
        name, pct = s.rsplit("(", 1)
        pct = pct.rstrip(")")
        return rf"\makecell{{{name.strip()}\\\textbf{{({pct})}}}}"
    return s

def save_latex(df, filename):
    lines = [" & ".join([""] + [cell_latex(c) for c in df.columns]) + r" \\ \hline"]
    for idx, row in df.iterrows():
        cells = [cell_latex(idx)] + [cell_latex(x) for x in row.values]
        lines.append(" & ".join(cells) + r" \\")
    (CONF_DIR / filename).write_text("\n".join(lines))

def save_png(df, filename, title):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.axis("off")
    tbl = ax.table(
        cellText=[[str(x).replace("(", "\n(") for x in row] for row in df.values],
        rowLabels=df.index, colLabels=df.columns, cellLoc='center', loc="center"
    )
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(7)
    tbl.scale(1, 1.4)
    ax.set_title(title, pad=12, fontsize=10)
    plt.tight_layout()
    fig.savefig(CONF_DIR / filename, dpi=300, bbox_inches="tight")
    plt.close(fig)

# 
#  BUILD MISS / NEAR-MISS TABLES
# 
cols      = [f"Top {i+1}" for i in range(TOP_K)]
miss_src  = pd.DataFrame(index=[m for m, *_ in models_and_arrays], columns=cols)
near_src  = miss_src.copy()

for model_name, pred, gt in models_and_arrays:
    ks_mat = compute_ks_matrix(pred, gt)              

    miss_share, near_share = contribution_per_kp(ks_mat)
    miss_idx = (-miss_share).argsort()[:TOP_K]
    near_idx = (-near_share).argsort()[:TOP_K]

    for rank, kp in enumerate(miss_idx):
        miss_src.loc[model_name, cols[rank]] = f"{keypoint_names[kp]} ({miss_share[kp]:.1f}%)"
    for rank, kp in enumerate(near_idx):
        near_src.loc[model_name, cols[rank]] = f"{keypoint_names[kp]} ({near_share[kp]:.1f}%)"

def build_rank_table(src_df):
    ranks = [str(i + 1) for i in range(TOP_K)]
    out   = pd.DataFrame(index=ranks, columns=src_df.index, dtype=object)
    for mdl in src_df.index:
        out[mdl] = src_df.loc[mdl, cols].values
    return out

miss_T = build_rank_table(miss_src)
near_T = build_rank_table(near_src)

# 
#  SAVE RESULTS
# 
save_latex(miss_T, "miss_top10_by_model.tex")
save_latex(near_T, "nearmiss_top10_by_model.tex")
save_png(miss_T, "miss_top10_by_model.png", "Top-10 Miss contribution")
save_png(near_T, "nearmiss_top10_by_model.png", "Top-10 Near-miss contribution")

print("Saved:")
print(" ", CONF_DIR / "miss_top10_by_model.tex")
print(" ", CONF_DIR / "nearmiss_top10_by_model.tex")
print(" ", CONF_DIR / "miss_top10_by_model.png")
print(" ", CONF_DIR / "nearmiss_top10_by_model.png")


```

### MISS NEAR MISS TABLES:

```python
import numpy as np
import pandas as pd
from pathlib import Path

# Constants
TOP_K = 5
NEAR_TH = 0.50  # Threshold for miss
GOOD_TH = 0.85  # Threshold for near miss
SIGMA = 0.025   # OKS sigma value

# Output directory
LATEX_DIR = Path(
    "/Users/marcusschou/Desktop/SDU/data_science/kurser_4/"
    "Speciale_Data_Science/INFERENCE_EVAL/eval_folder/latex_code/test_videos"
)
LATEX_DIR.mkdir(parents=True, exist_ok=True)

def align_by_frame(gt_arr, pred_arr):
    # Extract frame indices from GT and predictions
    gt_frame_idxs = gt_arr[:, 0, 3].astype(int)
    pred_frame_idxs = pred_arr[:, 0, 3].astype(int)
    common = np.intersect1d(gt_frame_idxs, pred_frame_idxs)

    if len(common) == 0:
        return None, None

    # Boolean masks and sorting indices to align frames
    gt_mask = np.isin(gt_frame_idxs, common)
    pred_mask = np.isin(pred_frame_idxs, common)
    sorted_gt_idx = np.argsort(gt_arr[gt_mask][:, 0, 3])
    sorted_pred_idx = np.argsort(pred_arr[pred_mask][:, 0, 3])

    # Return aligned GT and predictions
    aligned_gt = gt_arr[gt_mask][sorted_gt_idx][:, :, :2]
    aligned_pred = pred_arr[pred_mask][sorted_pred_idx][:, :, :2]
    return aligned_gt, aligned_pred

def compute_oks_matrix(gt, pred, sigma):
    F, K = gt.shape[:2]
    vis = ~np.isnan(gt[..., 0]) & ~np.isnan(gt[..., 1])  # GT visibility
    pred_valid = ~np.isnan(pred[..., 0]) & ~np.isnan(pred[..., 1])  # Prediction availability

    oks_mat = np.full((F, K), np.nan, dtype=np.float32)

    for f in range(F):
        for k in range(K):
            if not vis[f, k]:
                continue  # Skip invisible GT
            if not pred_valid[f, k]:
                oks_mat[f, k] = 0  # Visible GT but no prediction
                continue

            if k == K - 1:
                # Worm: use full bbox from all visible keypoints
                box_points = gt[f][vis[f]]
            else:
                # Other keypoints: exclude the worm (last keypoint)
                mask = vis[f].copy()
                mask[K - 1] = False
                box_points = gt[f][mask]

            area = compute_area(box_points)
            if area == 0:
                continue

            s = np.sqrt(area)
            d = np.linalg.norm(pred[f, k] - gt[f, k])
            ks = np.exp(-d**2 / (2 * s**2 * sigma**2))
            oks_mat[f, k] = ks

    return oks_mat

def compute_area(pts):
    x = pts[:, 0]
    y = pts[:, 1]
    if len(pts) < 2:
        return 0
    min_x, max_x = np.nanmin(x), np.nanmax(x)
    min_y, max_y = np.nanmin(y), np.nanmax(y)
    return (max_x - min_x) * (max_y - min_y)

def contribution_per_kp(ks_mat, near_th=NEAR_TH, good_th=GOOD_TH):
    valid = ks_mat > 0
    miss = (ks_mat <= near_th) & valid
    near = (ks_mat > near_th) & (ks_mat < good_th) & valid
    miss_share = 100 * miss.sum(0) / miss.sum() if miss.sum() else np.zeros(miss.shape[1])
    near_share = 100 * near.sum(0) / near.sum() if near.sum() else np.zeros(near.shape[1])
    return miss_share, near_share

def build_rank_table(src_df):
    ranks = [str(i+1) for i in range(TOP_K)]
    out = pd.DataFrame(index=ranks, columns=src_df.index, dtype=object)
    for m in src_df.index:
        out[m] = src_df.loc[m, [f"Top {i+1}" for i in range(TOP_K)]].values
    return out

def cell_latex(cell: str | float) -> str:
    if pd.isna(cell):
        return ""
    s = str(cell).replace("%", r"\%")
    if "(" in s and ")" in s:
        name, pct = s.rsplit("(", 1)
        pct = pct.rstrip(")")
        return rf"\makecell{{{name.strip()}\\\textbf{{({pct})}}}}"
    else:
        return s

def save_latex(df, filename):
    lines = [" & ".join([""] + [cell_latex(c) for c in df.columns]) + r" \\ \hline"]
    for idx, row in df.iterrows():
        cells = [cell_latex(idx)] + [cell_latex(x) for x in row.values]
        lines.append(" & ".join(cells) + r" \\")
    (LATEX_DIR / filename).write_text("\n".join(lines))

# Main loop
cols = [f"Top {i+1}" for i in range(TOP_K)]
miss_src = pd.DataFrame(index=model_dicts.keys(), columns=cols)
near_src = miss_src.copy()

for model_name, preds in model_dicts.items():
    oks_all = []

    for vid_name, pred_arr in preds.items():
        if vid_name not in gt_data:
            continue
        gt_arr = gt_data[vid_name]
        aligned_gt, aligned_pred = align_by_frame(gt_arr, pred_arr)
        if aligned_gt is None or aligned_pred is None:
            continue
        oks_mat = compute_oks_matrix(aligned_gt, aligned_pred, sigma=SIGMA)
        oks_all.append(oks_mat)

    if not oks_all:
        continue

    ks_mat = np.concatenate(oks_all, axis=0)
    miss_sh, near_sh = contribution_per_kp(ks_mat)

    miss_idx = (-miss_sh).argsort()[:TOP_K]
    near_idx = (-near_sh).argsort()[:TOP_K]

    for i, kp in enumerate(miss_idx):
        miss_src.loc[model_name, cols[i]] = f"{keypoint_names[kp]} ({miss_sh[kp]:.1f}%)"
    for i, kp in enumerate(near_idx):
        near_src.loc[model_name, cols[i]] = f"{keypoint_names[kp]} ({near_sh[kp]:.1f}%)"

miss_T = build_rank_table(miss_src)
near_T = build_rank_table(near_src)

save_latex(miss_T,  "miss_top10_by_model.tex")
save_latex(near_T,  "nearmiss_top10_by_model.tex")
```

### Error distances (Pixels):

```python
import numpy as np, pandas as pd
from pathlib import Path

# 
#  parameters / paths
# 
LATEX_DIR = Path(
    "/Users/marcusschou/Desktop/SDU/data_science/kurser_4/"
    "Speciale_Data_Science/INFERENCE_EVAL/eval_folder/latex_code/test_videos"
)
LATEX_DIR.mkdir(parents=True, exist_ok=True)
OUTFILE = LATEX_DIR / "dist_percentiles_table_rows.tex"

# 
#  helpers
# 
def flat_pixel_errors(gt: np.ndarray, pr: np.ndarray) -> np.ndarray:
    vis_gt    = ~np.isnan(gt[..., 0]) & ~np.isnan(gt[..., 1])
    finite_pr = ~np.isnan(pr[..., 0]) & ~np.isnan(pr[..., 1])
    valid     = vis_gt & finite_pr
    if not np.any(valid):
        return np.empty(0)
    d = np.linalg.norm(pr[..., :2] - gt, axis=-1)
    return d[valid]

def per_model_percentiles(gt_data: dict, preds_dict: dict):
    """Return (p50, p75, p95) pixel error across ALL videos for one model."""
    errs = []
    for vid, pr in preds_dict.items():
        if vid not in gt_data:
            continue
        # align frames
        gf = gt_data[vid][:, 0, 3].astype(int)
        pf = pr[:,      0, 3].astype(int)
        common = np.intersect1d(gf, pf)
        if not common.size:
            continue
        gmask, pmask = np.isin(gf, common), np.isin(pf, common)
        sgt = np.argsort(gt_data[vid][gmask][:, 0, 3])
        spr = np.argsort(pr[pmask][:, 0, 3])
        gt_al = gt_data[vid][gmask][sgt][:, :, :2]
        pr_al = pr[pmask][spr][:, :, :3]
        errs.append(flat_pixel_errors(gt_al, pr_al))
    if not errs:
        return np.nan, np.nan, np.nan
    all_err = np.concatenate(errs)
    return tuple(np.nanpercentile(all_err, q) for q in (50, 75, 95))

# 
#  models in short form + their prediction dictionaries
# 
models_and_preds = [
    ("SLEAP - Top-down model", preds_paths_slp_td),
    ("SLEAP - Single-instance model", preds_paths_slp_si),
    ("DeepLabCut - Bottom-up model", preds_paths_dlc_bu),
    ("DeepLabCut - Top-down model", preds_paths_dlc_td),
    ("LightningPose - Bottom-up model", preds_paths_lp_bu),
    ("LightningPose - TCN model", preds_paths_lp_tcn),
]

# 
#  build rows
# 
rows = []
for name, preds in models_and_preds:
    p50, p75, p95 = per_model_percentiles(gt_data, preds)
    rows.append({
        "Model": name,
        #  numbers only, no px 
        "p50": f"{p50:.2f}" if not np.isnan(p50) else "--",
        "p75": f"{p75:.2f}" if not np.isnan(p75) else "--",
        "p95": f"{p95:.2f}" if not np.isnan(p95) else "--",
    })

df = pd.DataFrame(rows)

# 
#  write LaTeX table rows
# 
with OUTFILE.open("w") as fh:
    fh.write("Model & dist.p50 & dist.p75 & dist.p95 \\\\ \\hline\n")
    for _, r in df.iterrows():
        fh.write(f"{r['Model']} & {r['p50']} & {r['p75']} & {r['p95']} \\\\\n")

print("LaTeX rows saved to:", OUTFILE)

```

### Jitter metric:

```python
import numpy as np
import pandas as pd
from pathlib import Path

tex_filename = "jitter_weighted_video_mean_std_with_paircount.tex"
LATEX_DIR = Path(
    "/Users/marcusschou/Desktop/SDU/data_science/kurser_4/"
    "Speciale_Data_Science/INFERENCE_EVAL/eval_folder/latex_code/test_videos"
)
LATEX_DIR.mkdir(parents=True, exist_ok=True)
LATEX_OUT = LATEX_DIR / tex_filename

model_names = ["SLEAP-td", "SLEAP-si", "DLC-td", "DLC-bu", "LP-bu", "LP-tcn"]

def jitter_per_video(arr):
    xy = arr[..., :2]
    frame_idx = arr[:, 0, 3].astype(int)
    consecutive = np.diff(frame_idx) == 1

    valid_now = ~np.isnan(xy[..., 0]) & ~np.isnan(xy[..., 1])
    valid_next = valid_now[:-1] & valid_now[1:]

    disp = np.linalg.norm(np.diff(xy, axis=0), axis=-1)
    disp[~valid_next | ~consecutive[:, None]] = np.nan

    valid_disp = disp[~np.isnan(disp)]
    return np.mean(valid_disp), valid_disp.size

summary_rows = []

for name, (mdl_key, preds) in zip(model_names, model_dicts.items()):
    per_video_means = []
    per_video_weights = []

    for arr in preds.values():
        mean_jitter, count = jitter_per_video(arr)
        if not np.isnan(mean_jitter) and count > 0:
            per_video_means.append(mean_jitter)
            per_video_weights.append(count)

    if per_video_means:
        weights = np.array(per_video_weights)
        means = np.array(per_video_means)

        weighted_mean = np.average(means, weights=weights)
        variance = np.average((means - weighted_mean)**2, weights=weights)
        weighted_std = np.sqrt(variance)

        row = {
            "Model": name,
            "Jitter (mean  std)": f"{weighted_mean:.2f}  {weighted_std:.2f}",
            "# Valid Pairs": int(np.sum(weights))
        }
    else:
        row = {
            "Model": name,
            "Jitter (mean  std)": "NaN  NaN",
            "# Valid Pairs": 0
        }

    summary_rows.append(row)

summary_df = pd.DataFrame(summary_rows)

with LATEX_OUT.open("w") as fh:
    header = "Model & Jitter (mean $\\pm$ std across videos) & \\# Valid Pairs \\\\ \\hline"
    fh.write(header + "\n")
    for _, row in summary_df.iterrows():
        line = f"{row['Model']} & {row['Jitter (mean  std)']} & {row['# Valid Pairs']} \\\\"
        fh.write(line + "\n")

print("LaTeX summary table saved to:", LATEX_OUT)

```
